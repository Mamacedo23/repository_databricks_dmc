```python
from pyspark.sql.functions import when, coalesce, to_date, substring, initcap, trim, col, lit, lower, current_timestamp, upper
from pyspark.sql.types import LongType, StringType, DateType
```

```python
catalog_name = "webinar"

df = (
    spark.table(f"{catalog_name}.bronze.tiendas")
    .dropDuplicates(["id_tienda"])
    .withColumn("updated_at", current_timestamp())
    .select(
        col("id_tienda").cast(LongType()),
        initcap(trim(col("nombre"))).alias("nombre"),
        initcap(trim(col("ciudad"))).alias("ciudad"),
        upper(trim(col("region"))).alias("region"),
        col("updated_at")
    )
)

df.write.format("delta").mode("overwrite").saveAsTable(f"{catalog_name}.silver.tiendas")
```
