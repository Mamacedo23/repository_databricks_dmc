```python
from pyspark.sql.functions import current_timestamp, col, hash, abs
```

```python
catalog_name = "webinar"
```

```python
df = (
    spark.table(f"{catalog_name}.silver.tiendas")
    .withColumn(
        "sk_tienda",
        abs(hash(col("id_tienda")))
    )
    .withColumn("updated_at", current_timestamp())
    .select(
        "sk_tienda",
        "id_tienda",
        "nombre",
        "ciudad",
        "region",
        "updated_at"
    )
)

df.write.format("delta").mode("overwrite").saveAsTable(f"{catalog_name}.gold.tiendas")
```
