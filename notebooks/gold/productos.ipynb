```python
from pyspark.sql.functions import current_timestamp, col, hash, abs
```

```python
catalog_name = "webinar"
```

```python
df = (
    spark.table(f"{catalog_name}.silver.productos")
    .withColumn(
        "sk_producto",
        abs(hash(col("id_producto")))
    )
    .withColumn("updated_at", current_timestamp())
    .select(
        "sk_producto",
        "id_producto",
        "nombre",
        "categoria",
        "precio",
        "updated_at"
    )
)

df.write.format("delta").mode("overwrite").saveAsTable(f"{catalog_name}.gold.productos")
```
